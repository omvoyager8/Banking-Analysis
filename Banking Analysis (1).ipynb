# Exploratory Data Analysis on Banking Dataset using Python

Column Heads:__

-- __ClientID__: The number assignged to each client.

-- __Name__: Name of Account Holder.

-- __age__: The age of the user.

-- __LocationID__: Location of user house.

-- __Joined Bank__: Date of joining the bank.

-- __Banking Contact__: Bank managers contact number.

-- __Nationality__: Nationality of user.

-- __Occupation__: The occupation of account holder.

---- __Fee Structure__: The fee structure based on high low and medium.

-- __Loyalty Classification__: Loyality classification based on relationship in case of trust,legal,seniority like gold,silver,platinum,jade.

-- __Estimated Income__: Estimated income of a account holder.

-- __Superannuation Savings__: The savings of account holders for after retirment life.

-- __Amount of Credit Cards__: Total number of credit cards.

-- __Credit Card Balance__: Credit card balance.

-- __Bank Loans__: Bank loan amount taken by member.

-- __Bank Deposits__: The bank deposit amount of user .

-- __Checking Accounts__: The checking accounts balance.

---- __Saving Accounts__: The savings account balance.

-- __Foreign Currency Account Business__: Specifies foreign currency of user.

-- __Properties Owned__: The properties hold by load owner or bank account holder.

-- __Risk Weighting__: Risk weighting values from 1 to 5.

-- __BRId__: Branch id of bank

-- __GenderId__: gender id for male 1 and female 2.

-- __IAId__: The values of internal audits.

```python
# replacing numpy null in place of NaN,Nan,None
df['Loyalty Classification'].replace(['NaN', 'Nan', 'None', ''], np.nan, inplace=True)
```

```python
df['Nationality'].unique()
```

```python
df['Nationality'] = df['Nationality'].str.title()
```

```python
df['Nationality'].unique()
```

```python
#cleaning column 'Estimated Income' and changing its dtype to int
df['Estimated Income'] = df['Estimated Income'].str.replace(',', '').str.strip()
```

```python
df['Estimated Income']=df['Estimated Income'].replace({
    'unknown':'50000'
})
```

```python
df['Estimated Income']=df['Estimated Income'].astype('float')
```

```python
df['Estimated Income']=df['Estimated Income'].astype('int')
```

```python
df['Fee Structure'].unique()
```

```python
# removing error values from column 'Fee Structure'
df['Fee Structure']=df['Fee Structure'].str.strip().str.lower()
```

```python
df['Fee Structure']=df['Fee Structure'].replace({
    'standard':'standard',
    'nan': 'standard'
})
```

```python
df['Fee Structure']=df['Fee Structure'].str.upper()
```

```python
df['Fee Structure'].unique()
```

```python
# cleaning column age 
df['Age']=df['Age'].str.strip()
```

```python
df['Age']=df['Age'].replace({
    'twenty-five':'25.0'
})
```

```python
# changing datatype of age to float and then to int 
df['Age']=df['Age'].astype(float)
```

```python
df['Age']=df['Age'].astype(int)
```

# Feature Engineering

```python
#creating new column 'High_Risk_Flag'

df['High_Risk_Flag'] = ((df['Risk Weighting'] >= 3) & (df['Bank Loans'] > 700000) & (df['Business Lending'] > 500000)).astype(int)
```

```python
df
```

```python
#creating new column age group
df['Age Group'] = pd.cut(df['Age'], bins=[18, 30, 45, 60, 100], labels=['Young', 'Adult', 'Senior', 'Elder'])
```

```python
df
```

```python
df.dtypes
```

```python
#creating new column 'Total Wealth'

df['Total Wealth'] = df['Estimated Income'] + df['Superannuation Savings'] + df['Bank Deposits'] + (df['Properties Owned'] * 100000)
```

```python
df
```

# Data Visualization

```python
sns.barplot(x='Loyalty Classification', y='Total Wealth', data=df)
```

```python
# Create age bins
df['Age Bin'] = pd.cut(df['Age'], bins=[18, 30, 45, 60, 100], labels=['18-30', '31-45', '46-60', '60+'])

# Plot value counts for each age bin
df['Age Bin'].value_counts().sort_index().plot(kind='bar', color='purple')

plt.title('Distribution of Wealth by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
df['Total Wealth'].plot(kind='hist', bins=[0, 100000, 200000, 300000, 400000, 500000], edgecolor='black')
title()
```

```python
df['Deposit Bin']=pd.cut(df['Bank Deposits'], bins=[0,100000,500000,1000000,1500000,2000000],labels=['0L-1L','1L-5L','5L-10L','10L-15L','15L-20L'])
plt.figure(figsize=(5, 6))
sns.scatterplot(x='Deposit Bin', y='Age Group', hue='Risk Weighting', data=df, palette='coolwarm')

plt.title('Age Group vs Bank Deposits (Colored by Risk)')
plt.tight_layout()
plt.show()
```
